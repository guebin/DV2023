# `#` 비대면응시 관련 세부지침

## A. 준비물

`-` 컴퓨터, 노트북, 아이패드 등: 시험지 확인 및 문제풀이 용도

`-` 핸드폰: Zoom을 통하여 주변상황을 및 컴퓨터 화면을 촬영하는 용도 +
학생증 촬영 용도

> Note: 중간에 핸드폰 및 노트북이 꺼지지 않도록 배터리 충전기를
> 준비한다.

`-` 학생증: 본인 인증 용도 (대리시험 방지)

> **Important**
>
> 학생증이 없는 경우 신분증 (주민등록증/여권/운전면허증) 으로
> 대체가능함.

## B. 시험전 준비사항 (16:00~16:20)

`-` LMS $\to$ 강의대화 $\to$ Zoom 화상강의 바로 가기로 입장

-   핸드폰, 컴퓨터 모두 입장 할 것
-   핸드폰은 주변상황 촬영용도 + 학생증 촬영용도
-   컴퓨터는 질문을 위한 용도

`-` 줌의 대화명은 소속,이름,기기,학번을 순서대로 적는다. (예시:
전북대_최규빈_핸드폰_202143052)

> Note: 동명이인이 있을 수 있으므로 학번을 같이 적으세요

`-` 핸드폰을 아래와 같이 배치하여 학생의 컴퓨터 화면 및 주변상황이
보이도록 한다.

![](https://github.com/guebin/2021IR/blob/master/_notebooks/img.png?raw=true)

> Note: 적절한 각도를 설정하기 어려운 경우 주변환경보다 컴퓨터의 화면이
> 잘 보이도록 설정할 것

`-` 학생증을 준비하여 시험 시작 직전에 본인의 얼굴과 학생증을 함께
촬영한다. (5초간)

## C. 시험시간 중 유의사항 (16:20~17:50)

`-` 질문은 줌의 채팅기능을 이용한다. (하울링 방지를 위해 육성질문은 하지
않음)

`-` Zoom에서 스피커 음소거를 하지 않는다. (전체 공지사항등이 있을때
음성으로 공지함)

`-` 핸드폰으로 Zoom참가 중 전화가 오면 거절하고 받지 않는다. (전화통화시

`-` 카카오톡 등 메신저앱을 공유하여 코드공유를 할 경우 F 처리함.
