`-` Line2d 오브젝트를 쓰지 않는 방법

```python
fig3 = plt.Figure()
ax3 = fig3.add_axes([0.125, 0.11, 0.775, 0.77])
ax3.plot([1,2,3,2],'--o')
fig3
```

`-` add_axes()를 쓰지 않는 방법 

```python
fig4 = plt.Figure()
ax4 = fig4.subplots()
ax4.plot([1,2,3,2],'--o')
fig4
```

`-` 더 생략하는 방법

```python
fig5, ax5 = plt.subplots(1)
ax5.plot([1,2,3,2],'--o')
```

## 정리 ($\star\star\star$)

`-` 결국 아래는 모두 같은 코드이다. 

```python
## 코드1 
plt.plot([1,2,3,2],'--o')
```

```python
## 코드2
fig,ax = plt.subplots()
ax.plot([1,2,3,2],'--o')
```

```python
## 코드3
fig = plt.Figure()
ax = fig.subplots()
ax.plot([1,2,3,2],'--o')
fig
```

```python
## 코드4
fig = plt.Figure()
ax = fig.add_axes([0.125, 0.11, 0.775, 0.77])
ax.plot([1,2,3,2],'--o')
fig
```

```python
## 코드5
fig = plt.Figure()
ax = fig.add_axes([0.125, 0.11, 0.775, 0.77])
ax.set_xlim([-0.15, 3.15])
ax.set_ylim([0.9, 3.1])
line = matplotlib.lines.Line2D(
    xdata=[0,1,2,3], 
    ydata=[1,2,3,2], 
    linestyle='--', 
    marker='o'
)
ax.add_line(line)
fig
```
